```python
import matplotlib.pyplot as plt
```

```python
# Saving model results as tuple of (Model Name, Dataset, AUC, F1, Accuracy)
# where dataset is "{feature set}_{dataset}"
data = [
    ("Neural Net", "1_Subset_no_weather", 0.6225, 0.5704, 0.5848),
    ("Neural Net", "1_Subset_weather", 0.6401, 0.5861, 0.5980),
    ("Neural Net", "1_All_balanced", 0.6849, 0.4936, 0.6656),
    ("Neural Net", "1 & 2_Subset_no_weather", 0.9578, 0.8991, 0.9026),
    ("Neural Net", "1 & 2_Subset_weather", 0.9579, 0.8987, 0.9021),
    ("Neural Net", "1 & 2_All_balanced", 0.9644, 0.8926, 0.9184),
    ("Random Forest", "1_Subset", 0.6889, 0.6303, 0.6368),
    ("Random Forest", "1_Subset", 0.6534, 0.6022, 0.6108),
    ("Random Forest", "1_All_balanced", 0.7372, 0.5831, 0.6916),
    ("Random Forest", "1 & 2_Subset_no_weather", 0.9610, 0.9029, 0.9057),
    ("Random Forest", "1 & 2_Subset_weather", 0.9633, 0.9027, 0.9057),
    ("Random Forest", "1 & 2_All_balanced", 0.9693, 0.8979, 0.9208),
    ("XGBoost Classifier", "1 & 2_Subset", 0.9683, 0.9118, 0.9135),
    ("XGBoost Classifier", "1_Subset", 0.6919, 0.6398, 0.6334),
    ("XGBoost Classifier", "1_Full Dataset", 0.7380, 0.5471, 0.6931)
]
```

```python
model_colors = {
    "Neural Net": "#1f77b4",  
    "Random Forest": "#ff7f0e",  
    "XGBoost Classifier": "#2ca02c"  
}
```

```python
# Unpacking the sorted data
# model_dataset_labels = [f"{model}_{dataset}" for model, dataset, _ in data]

auc_values = [auc for _, _, auc, _, _ in data]
f1_values = [f1 for _, _, _, f1, _ in data]
accuracy_values = [accuracy for _, _, _, _, accuracy in data]

colors = [model_colors[model] for model, _, _, _, _ in data]
```

```python
plt.figure(figsize=(5, 5))

# Plot each model group with its corresponding color
for model, color in model_colors.items():
    # Select the points corresponding to the current model
    indices = [i for i, m in enumerate(data) if m[0] == model]
    plt.scatter([accuracy_values[i] for i in indices], 
                [auc_values[i] for i in indices], 
                color=color, 
                label=model)

plt.xlabel('Accuracy')
plt.ylabel('AUC')
plt.title('Model Accuracies by Dataset and Feature Set (Sorted by Accuracy)')

# Create the legend
plt.legend()

plt.tight_layout()
plt.show()
```
